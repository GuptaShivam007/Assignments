```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('company_sales_data.csv')
data.head()
```

```python
y1=data.iloc[:,1].values
y2=data.iloc[:,2].values
x=data.iloc[:,0].values
```

```python
import seaborn as sns
```

```python
sns.barplot(x=x,y=y1,label='facecream')
sns.barplot(x=x,y=y2,label='facewash')
plt.title('Product Data')
plt.xlabel('Month')
plt.ylabel('Units')
plt.show()
```

```python
sns.barplot(x=x,y=y1)
plt.title('Facecream Sales Data')
plt.xlabel('Months')
plt.ylabel('Units')
plt.show()
```

```python
sns.barplot(x=x,y=y2)
plt.title('Facewash Sales Data')
plt.xlabel('Months')
plt.ylabel('Units')
plt.show()
```
