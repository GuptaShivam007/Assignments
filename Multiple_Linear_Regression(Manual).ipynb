```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('Training set.csv')
x1=dataset.iloc[:,0].values
x2=dataset.iloc[:,1].values
y=dataset.iloc[:,-1].values
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
y=le.fit_transform(y)
```

```python
x1=list(x1)
x2=list(x2)
y=list(y)
```

```python
x1
```

```python
sumx1=0
sqx1=0
for i in x1:
  sumx1=sumx1+i
  sqx1=sqx1+i*i
```

```python
sumx2=0
sqx2=0
for i in x2:
  sumx2=sumx2+i
  sqx2=sqx2+i*i
```

```python
len(x1)
```

```python
sumx1y=0
i=0
while(i<len(x1)):

    sumx1y=sumx1y+x1[i]*y[i]
    i+=1

```

```python
sumx2y=0
i=0
while(i<len(x2)):

    sumx2y=sumx2y+x2[i]*y[i]
    i+=1
```

```python
sumy=0
for i in y:
  sumy=sumy+i
```

```python
sumx1x2=0
i=0
while(i<len(x1)):

    sumx1x2=sumx1x2+x1[i]*x2[i]
    i+=1
```

```python
sqX1=(sqx1-(sumx1**2))/len(x1)
sqX2=(sqx2-(sumx2**2))/len(x2)
sumX1Y=(sumx1y-(sumx1*sumy))/len(x1)
sumX2Y=(sumx2y-(sumx2*sumy))/len(x2)
sumX1X2=(sumx1x2-(sumx2*sumx1))/len(x2)
```

```python
b1=((sqX2*sumX1Y)-(sumX1X2*sumX2Y))/((sqX1*sqX2)-(sumX1X2**2))
b2=((sqX1*sumX2Y)-(sumX1X2*sumX1Y))/((sqX1*sqX2)-(sumX1X2**2))
b0=sumy/len(y)-b1*sumx1/len(x1)-b2*sumx2/len(x2)
```

```python
ans=[]
i=0
while(i<len(x1)):
  ansy=b0+b1*x1[i]+b2*x2[i]
  ans.append(ansy)
  i+=1
```

```python
y=np.array(y)
ans=np.array(ans)
```

```python
from sklearn.metrics import r2_score
```

```python
x=dataset.iloc[:,:-1]
```

```python
plt.plot(x,y,color='red')
plt.plot(x,ans,color='blue')
plt.show()
```

```python
r2_score(y,ans)
```
